```python
# 1: print_board(board)
def print_board(board):
    for row in board:
        print(" | ".join(row))
    print()
# 2: check_winner(board)
def check_winner(board):
    # Check rows and columns
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != " ":
            return board[i][0]  # Row win
        if board[0][i] == board[1][i] == board[2][i] != " ":
            return board[0][i]  # Column win
    # Check diagonals
    if board[0][0] == board[1][1] == board[2][2] != " ":
        return board[0][0]  # Main diagonal
    if board[0][2] == board[1][1] == board[2][0] != " ":
        return board[0][2]  # Anti-diagonal
    return None
#Step 3: is_moves_le (board)
def is_moves_left(board):
    return any(" " in row for row in board)
#Step 4: minimax(board, depth, is_maximizing)
def minimax(board, depth, is_maximizing):
    winner = check_winner(board)
    if winner == "X":
        return 10 - depth  # AI wins
    if winner == "O":
        return depth - 10  # Player wins
    if not is_moves_left(board):
        return 0  # Draw

#Step 5 : find_best_move(board)
#AI uses minimax() to choose the best possible move.
def find_best_move(board):
    best_score = -math.inf
    best_move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if board[i][j] == " ":
                board[i][j] = "X"  # Try move
                score = minimax(board, 0, False)  # Simulate opponent's turn
                board[i][j] = " "  # Undo move
                if score > best_score:
                    best_score = score
                    best_move = (i, j)
    return best_move
#Step 6: play_ c_tac_toe()
#This is the main game loop.
def play_tic_tac_toe():
    board = [[" " for _ in range(3)] for _ in range(3)]
    print("You are O, AI is X. Enter moves as row and column (0, 1, or 2).")

    while True:
        print_board(board)

        # Human move
        while True:
            try:
                row = int(input("Your move row (0-2): "))
                col = int(input("Your move col (0-2): "))
                if board[row][col] == " ":
                    board[row][col] = "O"
                    break
                else:
                    print("Cell occupied, try again.")
            except:
                print("Invalid input, try again.")
#Asks the player for a move and ensures the input is valid and the cell is empty.


        if check_winner(board) == "O":
            print_board(board)
            print("You win!")
            break
        if not is_moves_left(board):
            print_board(board)
            print("It's a draw!")
            break
#Checks if the player has won or if the board is full.

        # AI move
        ai_move = find_best_move(board)
        board[ai_move[0]][ai_move[1]] = "X"
        print("AI plays:", ai_move)

        if check_winner(board) == "X":
            print_board(board)
            print("AI wins!")
            break
        if not is_moves_left(board):
            print_board(board)
            print("It's a draw!")
            break
```
