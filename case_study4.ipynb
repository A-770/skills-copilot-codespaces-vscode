```python
# To-Do Notes Builder
to_do_list = []

def display_menu():
    print("\nTo-Do Notes Builder")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Delete Task")
    print("4. Mark Task as Completed")
    print("5. Exit")

def add_task():
    task = input("\nEnter your task: ")
    to_do_list.append({"task": task, "completed": False})
    print("Task added successfully!")

def view_tasks():
    print("\nYour To-Do List:")
    if not to_do_list:
        print("No tasks available.")
    else:
        for i, item in enumerate(to_do_list, 1):
            status = "[✔]" if item["completed"] else "[ ]"
            print(f"{i}. {status} {item['task']}")

def delete_task():
    view_tasks()
    try:
        task_number = int(input("\nEnter task number to delete: "))
        if 1 <= task_number <= len(to_do_list):
            to_do_list.pop(task_number - 1)
            print("Task deleted successfully!")
        else:
            print("Invalid task number!")
    except ValueError:
        print("Please enter a valid number.")

def mark_task_completed():
    view_tasks()
    try:
        task_number = int(input("\nEnter task number to mark as completed: "))
        if 1 <= task_number <= len(to_do_list):
            to_do_list[task_number - 1]["completed"] = True
            print("Task marked as completed!")
        else:
            print("Invalid task number!")
    except ValueError:
        print("Please enter a valid number.")

# Main loop
while True:
    display_menu()
    try:
        choice = int(input("\nEnter your choice: "))
        if choice == 1:
            add_task()
        elif choice == 2:
            view_tasks()
        elif choice == 3:
            delete_task()
        elif choice == 4:
            mark_task_completed()
        elif choice == 5:
            print("\nExiting... Have a great day!")
            break
        else:
            print("Invalid choice, please select from 1-5.")
    except ValueError:
        print("Please enter a valid number.")
```
